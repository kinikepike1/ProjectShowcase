```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Ransomware Attacks.csv")
```

```python
data.sector.replace(['tech ','telecomms','finace'],['tech','telecoms','finance'], inplace=True)
```

```python
targeted_sectors = {}
for sector in data.sector.unique():
    try:
        count = data['sector'].value_counts()[sector]
        targeted_sectors[sector] = count
    except:
        continue
```

```python
fig, bar = plt.subplots(figsize = (10, 6))
bar.barh(list(targeted_sectors.keys()),list(targeted_sectors.values()), color = 'red', edgecolor = 'darkred')
bar.set(xlabel = 'Times Targeted',title = 'Business Sectors of Ransomware Attacks')
plt.show()
```

```python
target_data = data.query("sector in ['tech','healthcare','government','academic']") 
org_size = {}
for org in target_data['organisation size 1,5,10,25,100,300 '].unique():
    try:
        count = target_data['organisation size 1,5,10,25,100,300 '].value_counts()[org]
        org_size[org] = count
    except:
        continue
```

```python
fig, pie = plt.subplots(figsize = (10, 6))
pie.pie(list(org_size.values()), labels=list(org_size.keys()))
plt.show()
```

```python
emp_data = data.dropna(subset=["no of employees"])
emp_count = {}
for emp in emp_data["no of employees"].unique():
    try:
        count = emp_data['no of employees'].value_counts()[emp]
        emp_count[emp] = count
    except:
        continue
        
fig, bar = plt.subplots(figsize = (10, 6))
bar.barh(list(emp_count.keys()),list(emp_count.values()), color = 'cyan', edgecolor = 'darkblue')
bar.set(xlabel = 'Times Targeted',ylabel='# of Employees')
plt.show()
```

```python
fig, scat = plt.subplots(figsize = (10, 9))
scat.scatter(data['ransom cost'],data['revenue $USD million '], alpha=.5)
scat.set(xlabel='Amount Ransomed ($mil)',ylabel='Revenue ($mil)')
plt.show()
```
